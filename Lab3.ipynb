## Problem 1

The unit step function

$ u(t) = \begin{cases}
1 & t \geq 0 \\
0 & t < 0 
\end{cases}$

can be coded using numerical Python as given below.

Sketch a graph of the unit step function $f = u(t)$ and annotate the graph.
- Title: "Unit step function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines

Notice that the values for **t** are already given and you can use them to calculate the values of **f**.

```python
import numpy as np
import matplotlib.pyplot as plt
def step(t):
    u = 1.0*(t>=0)
    return u

t = np.arange(-1, 5, 0.01)
f = step(t)

plt.plot(t,f)
plt.grid(True)
plt.title("Unit step function")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

## Problem 3

Rectangular pulse can be defined by subtrackting two delayed unit step functions. Sketch a graph of a rectangular pulse function $f = u(t-1) - u(t-3)$. Annotate the graph.
- Title: "Rectangular pulse function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines

Notice that the array for t is already given and you can use that to calculate the values of **f**.

```python
import numpy as np
import matplotlib.pyplot as plt
def step(t):
    u = 1.0*(t>=0)
    return u

t = np.arange(-1, 5, 0.01)
f = step(t-1) -step(t-3)

plt.plot(t,f)
plt.grid(True)
plt.title("Rectangular pulse function")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

## Problem 4

By multiplying the rectangular pulse with another function, we can create pulse functions with arbitrary shape.

Sketch a graph of a polynomial shaped pulse function $f = (u(t-1) - u(t-3)) \cdot (-t^2 + 4t - 3)$. 

Annotate the graph.
- Title: "Polynomial pulse function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines

Notice that the array for t is already given and you can use that to calculate the values of **f**.

```python
import numpy as np
import matplotlib.pyplot as plt
def step(t):
    u = 1.0*(t>=0)
    return u

t = np.arange(-1, 5, 0.01)
f = (step(t-1) -step(t-3))*(-t**2 + 4*t -3)

plt.plot(t,f)
plt.grid(True)
plt.title("Polynomial pulse function")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

## Problem 5

By using the rectangular pulse functions and multiplication (see previous problem) calculate the values for a piecewise continuous triangular function

$$f = \begin{cases}
t/2 & 0 \leq t < 2 \\
-t/2 + 2 & 2 \leq t < 4 \\
0 & \text{otherwise}
\end{cases}$$

Sketch a graph of the function and annotate the graph.
- Title: "Triangular pulse function"
- x-label: "Time (s)"
- y-label: "Amplitude"
- Add grid-lines

Notice that the array for t is already given and you can use that to calculate the values of **f**.

Tip: Create first two pulse functions $f_1 = t/2$, when $0 \leq t < 2$ and $f_2 = -t/2 + 2$, when $ 2 \leq t < 4$ and then sum them together.

```python
import numpy as np
import matplotlib.pyplot as plt
def step(t):
    u = 1.0*(t>=0)
    
    return u

t = np.arange(-1, 5, 0.01)
f1 = (step(t-0)-step(t-2)) * (t /2) 
f2 = (step(t-2)-step(t-4)) * (-t /2 + 2) 
f = f1 + f2

plt.title("Triangular pulse function")
plt.plot(t, f)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

           
           

```
