## Problem 1

Estimate the value of the following integrals

a) $S_1 = \int_0^5 \left( x^2 + 2x \right) dx$

b) $S_2 = \int_0^\pi sin(x) dx$

```python
# a)
x = np.linspace(0,5,50)
def f(x):
    return x**2 + 2*x
"""#print(x, f(x))
plt.plot(x, f(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()"""
S1 = 0
for i in np.arange(len(x) - 1):
    a = x[i]
    b = x[i+1]
    S1 = S1 + (b-a)*(f(a)+f(b))/2
print(f"S1= {S1:.1f}")

plt.plot(x, f(x))
plt.stem(x, f(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# b)
x = np.linspace(0,math.pi,50)
def g(x):
    return sin(x)
"""print(x, f(x))
plt.plot(x, f(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()"""
S2 = 0
for i in np.arange(len(x) - 1):
    a = x[i]
    b = x[i+1]
    S2 = S2 + (b-a)*(g(a)+g(b))/2
print(f"S2= {S2:.1f}")

plt.plot(x, g(x))
plt.stem(x, g(x))
plt.grid()
plt.xlabel('x')
plt.ylabel('g(x)')
plt.show()
```

## Problem 2

Estimate the position from 0 to 10 s, when the velocity is given by

a) $v_1(t)= -9.81 \cdot t$

b) $v_2(t) = -0.5t^2 + 5t$

Use time step $t_s = 0.01$

Plot both the velocity and the estimate position.

```python
# a)
ts = 0.01
t = np.arange(0, 10.01, ts)
v = -9.8*t
p = np.zeros(len(t))

for n in np.arange(len(t) - 1):
    p[n + 1] = p[n] + ts*(v[n + 1] + v[n])/2
    
plt.plot(t, v)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.show()
    
plt.plot(t, p)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.show()
```

```python
# b)
ts = 0.01
t = np.arange(0, 10.01, ts)
v = -0.5*t**2 + 5*t
p = np.zeros(len(t))
N = len(p)
for n in np.arange(N - 1):
    p[n + 1] = p[n] + ts*(v[n + 1] + v[n])/2
    
plt.plot(t, v)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.show()

plt.plot(t, p)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.show()
```

## Problem 4

Collect the accelerometer data using [Physics Toolbox Sensor Suite](https://play.google.com/store/apps/details?id=com.chrystianvieyra.physicstoolboxsuite&hl=en&gl=US) mobile application.

1. Place your mobile phone on a table (e.g. on a flat surface).
1. Repeate a figure (for example circle) with your mobile by moving it on the surface several times (5-10x).
1. Send the collected data to yourself by email. 
1. Estimate the 2D position (x and y coordinates) of your movement numerically using Python.

Draw plots of the acceleration, velocity and position data.

```python
from pathlib import Path
import pandas as pd


filename = 'circle3.csv'

file = Path(filename)
file.write_text(file.read_text().replace('-', '-'))

df = pd.read_csv(filename, 
                sep= ',',
                index_col = False)

#df.head(5)

ax = df['ax'].values
ay = df['ay'].values
t2 = df['time'].values


# Plot acceleration values
plt.plot(t2, ax, label='x-axis')
plt.plot(t2, ay, label='y-axis')
plt.grid()
plt.title('Acceleration')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
plt.legend()
plt.show()

#Velocity calculation
vx = np.zeros(len(t2))
vy = np.zeros(len(t2))

for n in np.arange(len(t2) - 1):
    vx[n + 1] = vx[n] + (t2[n+1]-t2[n]) * (ax[n + 1] + ax[n]) / 2
    vy[n + 1] = vy[n] + (t2[n+1]-t2[n]) * (ay[n + 1] + ay[n]) / 2


# Plot velocity values
plt.plot(t2, vx, label='x-axis')
plt.plot(t2, vy, label='y-axis')
plt.title('Integrated velocity of x and y axis')
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.show()

#Corrected Velocity
kx = vx[-1]/t2[-1]
ky = vy[-1]/t2[-1]
x_trend = kx*t2
y_trend = ky*t2
vx_correct = vx- x_trend
vy_correct = vy- y_trend

plt.plot(t2, vx_correct, label ='corrected x velocity')
plt.plot(t2, vy_correct, label ='corrected y velocity')
plt.title('Integrated and corrected velocity of x and y axis')
plt.grid()
plt.xlabel('Time(s)')
plt.ylabel('velocity(m/s)')
plt.legend()
plt.show()
```

```python
# Calculate position
x = np.zeros(len(t2))
y = np.zeros(len(t2))
for n in np.arange(len(t2) - 1):
    x[n + 1] = x[n] + (t2[n+1]-t2[n]) * (vx_correct[n + 1] + vx_correct[n]) / 2
    y[n + 1] = y[n] + (t2[n+1]-t2[n]) * (vy_correct[n + 1] + vy_correct[n]) / 2

# Plot position values
plt.plot(x, y)
plt.grid()
plt.xlabel('x position')
plt.ylabel('y position')
plt.show()
```
