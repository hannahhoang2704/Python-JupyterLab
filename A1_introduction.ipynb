## Import libraries and read data

More info about the dataset, read: [Iris flower data set](https://en.wikipedia.org/wiki/Iris_flower_data_set).

```python
# Import needed libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read the datafile and show the first 5 rows
file_name = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(file_name)
df.head()
```

## Problem 1

**Calculate the mean value of all data in column `sepal_length`.**

```python
mean = df['sepal_length'].mean()
print(f'Mean: {mean}')
```

**Calculate the standard deviation of all data in column `sepal_width`.**

```python
sd = df['sepal_width'].std()
print(f"Standard deviation: {sd}")
```

**Find the min and max values of all data in column `petal_length`.**

```python
min = df['petal_length'].min()
max = df['petal_length'].max()
print(f'Min values: {min}')
print(f'Max values: {max}')
```

**Find the 25%, 50%, and 75% percentiles for column `petal_width`.**

```python
# Percentiles
pw = df["petal_width"]
p = np.percentile(pw, [25, 50, 75])
print(f'25% = {p[0]}')
print(f'50% = {p[1]}')
print(f'75% = {p[2]}')
```

## Problem 3

**Draw a stem-plot of the values in column `sepal_length`**.

```python
# stem-plot of values in column sepal_length
sl = df['sepal_length']
plt.figure(figsize=(5,8))
plt.stem(sl, orientation = 'horizontal', bottom = 5.843)
plt.title('Stem-plot of sepal_length values')
plt.xlabel('Value of sepal_length')
plt.ylabel('Count')
plt.show()
```

**Draw a histogram of the values in column `sepal_width`.**

```python
# Histrogram of the values in column sepal_width
sw = df['sepal_width']
plt.hist(sw, bins = 10)
plt.xlabel('Value of sepal_width')
plt.ylabel('Count')
plt.title('Histogram of sepal_width values')
plt.show()
```

**Draw a violing plot of the values in column `petal_length`.**

```python
# Violin-plot of petal_length
pl = df['petal_length']
plt.violinplot(pl)
plt.title('Violin-plot of petal_length values')
plt.ylabel('Value of petal_length')
plt.show()
```

**Draw a scatter-plot of the values in column `petal_width` (in y-axis) vs. `sepal_width` (in x-axis)**.

```python
#Scatter-plot of values in column petal_width (in y-axis) vs. sepal_width (in x-axis)
pw = df['petal_width']
sw = df['sepal_width']
plt.scatter(sw, pw)
plt.title('Scatter plot of petal_width vs. sepal_width')
plt.xlabel('Value of sepal_width')
plt.ylabel('Value of petal_width')
plt.grid()
plt.show()
```
